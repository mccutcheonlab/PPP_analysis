```python
### Reads in data
pickle_in = open('..\\data\\ppp_dfs_latencies.pickle', 'rb')
df_nr_cas, df_nr_malt, df_pr_cas, df_pr_malt = pd.read_pickle(pickle_in, compression=None)
```

### Panel A - non-restricted correlations

```python
fig4A = scatter_plus_density(df_nr_cas["latency"], df_nr_cas["time2peak"], df_nr_malt["latency"], df_nr_malt["time2peak"])
```

### Panel B - non-restricted correlations

```python
fig4B = scatter_plus_density(df_pr_cas["latency"], df_pr_cas["time2peak"], df_pr_malt["latency"], df_pr_malt["time2peak"], colors=["blue", "green"])
```

```python
u, p = mannwhitneyu(df_nr_cas["latency"], df_nr_malt["latency"], alternative="two-sided")
print("NR rats: Lick latency, Casein vs. Malt - U={}, p={}".format(u, p))

u, p = mannwhitneyu(df_nr_cas["time2peak"], df_nr_malt["time2peak"], alternative="two-sided")
print("NR rats: Time to peak, Casein vs. Malt - U={}, p={}".format(u, p))

u, p = mannwhitneyu(df_pr_cas["latency"], df_pr_malt["latency"], alternative="two-sided")
print("PR rats: Lick latency, Casein vs. Malt - U={}, p={}".format(u, p))

u, p = mannwhitneyu(df_pr_cas["time2peak"], df_pr_malt["time2peak"], alternative="two-sided")
print("PR rats: Time to peak, Casein vs. Malt - U={}, p={}".format(u, p))
```
